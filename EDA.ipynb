```python
#Build function to recode categorical variables with many categories
def recode_categorical(dataframe, feature):
  top_categories = dataframe[feature].value_counts().nlargest(5).index.tolist()
  dataframe[feature] = dataframe[feature].where(dataframe[feature].isin(top_categories), 'other')
  return dataframe

#Build function to one-hot encode categorical variables
def one_hot_encode(dataframe):
    #Select categorical variables
    categorical_vars = dataframe.select_dtypes(include=['object']).columns

    #Handle variables with many categories
    for var in categorical_vars:
      if dataframe[var].nunique() > 6:
        dataframe = recode_categorical(dataframe, var)

    #Encode the variables with dummies
    encoded_vars = pd.get_dummies(dataframe[categorical_vars], drop_first=True)
    #Add the encoded variables to dataframe
    dataframe = pd.concat([dataframe, encoded_vars], axis=1)
    #Drop original categorical variables
    dataframe = dataframe.drop(columns=categorical_vars)

    return dataframe

#Visualize variable distributions
def plot_hist(dataframe, feature):
    plt.figure(figsize=(8, 6))
    sns.histplot(dataframe[feature], bins=30, kde=False, color='green')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
